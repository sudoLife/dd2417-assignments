```python
from anytree import Node, RenderTree, DoubleStyle

def build_ast(expression):
    if isinstance(expression, list):
        operator = expression[0]
        operands = [build_ast(operand) for operand in expression[1:]]
        return Node(operator, children=operands)
    else:
        return Node(str(expression))

ast1 = build_ast(
    ['S', [
        'NP', [
            'Pron', ['He']
        ],
    ],
        [
        'VP', 
        [
            'Verb', 'built'
        ],
        [
            'NP',
            ['Det', 'the'],
            ['Noun', 'box'],
            [
                'PP',
                ['Prep', 'with'],
                ['NP',
                    ['Det', 'a'],
                    ['Noun', 'hammer'],
                    ['PP',
                     ['Prep', 'in'],
                     ['NP',
                      ['Det', 'the'],
                      ['Noun', 'Yard'],
                      ['PP',
                       ['Prep', 'behind'],
                       ['NP',
                        ['Det', 'the'],
                        ['Noun', 'house']
                        ]
                       ]
                      ],
                     ]
                 ]
            ]
        ]
    ]
    ]
)

export_tree(ast1, "ast1.png")


# for pre, _, node in RenderTree(ast1, style=DoubleStyle):
#     print("%s%s" % (pre, node.name))
```

```python
from anytree.exporter import UniqueDotExporter

def export_tree(tree, filename: str) -> None:
    def edgefunc(node, child):
        return '--'
    def determine_color(name: str):
        lookup = {
            'S': 'black',
            'NP': 'green',
            'VP': 'red',
            'PP': 'blue',
            'Det': 'turquoise',
            'Pron': 'brown',
            'Verb': 'pink',
            'Adj': 'coral',
            'Prep': 'orange',
            'Noun': 'violet'
        }

        if name in lookup:
            return lookup[name]
        return 'black'
    def nodeattrfunc(n): 
        return "label=\"{}\", fontcolor={}, shape=none".format(n.name, determine_color(n.name))
    UniqueDotExporter(tree, graph='graph', edgetypefunc=edgefunc, nodeattrfunc=nodeattrfunc).to_picture(filename)


# export_tree(ast1, 'ast1.png')
# export_tree(ast2, 'ast2.png')
```
